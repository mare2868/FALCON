Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
from math import radians, sin, cos, sqrt, atan2

# Function to calculate distance between two points using Haversine formula
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # radius of the Earth in kilometers
    
    # Convert latitude and longitude from degrees to radians
    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)
    
    # Calculate the change in coordinates
    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad
    
    # Calculate the Haversine distance
    a = sin(dlat / 2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    # Distance in kilometers
    distance = R * c
    return distance

# Initialize a map centered at NASA Johnson Space Center
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Iterate through each launch site to calculate distances to other proximities
for index, site in launch_sites_df.iterrows():
    site_name = site['Launch Site']
    site_lat = site['Lat']
    site_lon = site['Long']
    
    # Create a MarkerCluster for proximities
    marker_cluster_proximities = MarkerCluster(name=f"{site_name} Proximities").add_to(site_map)
    
    # Iterate through proximities to calculate distances and add markers
    for idx, launch in spacex_df.iterrows():
        launch_site_lat = launch['Lat']
        launch_site_lon = launch['Long']
        launch_site_name = launch['Launch Site']
        
        # Skip if the launch site is the same as the current site
        if site_name == launch_site_name:
            continue
        
        # Calculate distance using Haversine formula
        distance = haversine(site_lat, site_lon, launch_site_lat, launch_site_lon)
        
        # Create a popup text with distance information
        popup_text = f"Distance to {launch_site_name}: {distance:.2f} km"
        
        # Create a marker for each proximity with distance information
        folium.Marker(
            location=[launch_site_lat, launch_site_lon],
            popup=popup_text,
            icon=folium.Icon(color='purple', icon='cloud')
        ).add_to(marker_cluster_proximities)

# Add LayerControl to the map to toggle MarkerClusters
folium.LayerControl().add_to(site_map)

# Display the map
site_map
```

```python
import folium
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# Function to calculate distance between two points using Haversine formula
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # approximate radius of Earth in km

    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)

    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad

    a = sin(dlat / 2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Example coordinates (replace with actual coordinates found on the map)
launch_site_lat = 28.562302
launch_site_lon = -80.577356
closest_point_lat = 28.5615
closest_point_lon = -80.575

# Calculate distance between launch site and closest point of interest
distance_closest_point = calculate_distance(launch_site_lat, launch_site_lon, closest_point_lat, closest_point_lon)

# Create folium map centered around launch site
map_center = [launch_site_lat, launch_site_lon]
m = folium.Map(location=map_center, zoom_start=15)

# Add marker for launch site
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    popup='Launch Site',
    icon=folium.Icon(color='blue', icon='rocket')
).add_to(m)

# Add marker for closest point of interest (e.g., city, railway, highway)
folium.Marker(
    location=[closest_point_lat, closest_point_lon],
    popup='Closest Point of Interest',
    icon=folium.Icon(color='green', icon='info-sign')
).add_to(m)

# Create a PolyLine between launch site and closest point of interest
lines = folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [closest_point_lat, closest_point_lon]],
    weight=1,
    color='orange'
)
m.add_child(lines)

# Add marker to display distance
distance_marker = folium.Marker(
    location=[closest_point_lat, closest_point_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12; color:#d35400;"><b>{distance_closest_point:.2f} KM</b></div>'
    )
).add_to(m)

# Display the map
m
```
